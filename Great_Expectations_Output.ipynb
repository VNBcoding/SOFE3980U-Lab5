```python
import pandas as pd
import great_expectations as gx
```

```python
df = pd.read_csv("Labels.csv")
df.head()
```

```python
context = gx.get_context()
data_source = context.data_sources.add_pandas("pandas")
data_asset = data_source.add_dataframe_asset(name="labels_dataframe")
batch_definition = data_asset.add_batch_definition_whole_dataframe("batch_definition")
batch = batch_definition.get_batch(batch_parameters={"dataframe": df})
```

```python
expectation_1 = gx.expectations.ExpectColumnValuesToBeBetween(
    column="Timestamp", min_value=0
)
result_1 = batch.validate(expectation_1)
result_1
```

```python
expectation_2 = gx.expectations.ExpectColumnValuesToNotBeNull(
    column="pedestrianLocationX_TopLeft"
)
result_2 = batch.validate(expectation_2)
result_2
```

```python
expectation_3 = gx.expectations.ExpectColumnValuesToBeUnique(
    column="Occluded_Image_view"
)
result_3 = batch.validate(expectation_3)
result_3
```
